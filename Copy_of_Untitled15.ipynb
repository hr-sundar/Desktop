```python
import pandas as pd

data = pd.read_excel("/content/Grocery_Inventory_and_Sales_Dataset (1).xlsx")


category_counts = data.groupby('Catagory')['Product_ID'].count().reset_index()


top_3_categories = category_counts.sort_values(by='Product_ID', ascending=False).head(3)


top_3_products = data[data['Catagory'].isin(top_3_categories['Catagory'])]


print("Top 3 Categories with Products:")
print(top_3_products[['Catagory', 'Product_Name']])
```
